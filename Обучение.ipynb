```python
import sqlite3
# Создание базы данных в памяти
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
# Чтение содержимого файла
sql_file = 'example.sql'
with open(sql_file, 'r') as file:
  sql_script = file.read()
# Выполнение SQL-скрипта
cursor.executescript(sql_script)
conn.commit()
# Выполнение запроса для вывода данных
select_query = 'SELECT * FROM РАБОТНИКИ'
cursor.execute(select_query)
result = cursor.fetchall()
# Вывод данных на экран
for row in result:
  print(row)
# Закрытие соединения
conn.close()
```

```python
import pandas as pd
df=pd.DataFrame({'x':[1,3,2],'y':[2,4,1]})
df
```

Можно удалить строки в датафрейме - с помощью команды drop, (параметр inplace отвечает за то, что строки будут удалены полностью (команда True) или просто отключена их видимость (False))

```python
df.drop(1, axis=0, inplace=True)
df
```

Для удаления столбца используется команда del: (за # написаны альтернативный вариант удаленя столбца. Стоит учитывать что команда drop используется для библиотеки Pandas, а del - для Python)

```python
del df['x'] # df.drop('x', axis=1)
df
```

Сравнение датафреймов

```python
df1 = pd.DataFrame({'x':[1,3,2], 'y':[2,4,1]})
df2 = pd.DataFrame({'x':[3,1,2], 'y':[0,2,2]})
df1
```

```python
df2
```

```python
df1>=df2
```

Можно использовать составные условия. В этом случае
**операторы сравнения «и» и «или» задаются символами «&» и «|»**
соответственно:

```python
df1[df1 >1]
```

```python
df2[(df2 < 2) & (df2 >0)]
```

```python
df2[(df2 < 1) | (df2 >2)]
```

**df['column_name'].sum()**

**.mean()** – расчет среднего значения по столбцу:

**df['column_name'].mean()**

**.median()** – расчет медианы по столбцу:

**df['column_name'].median()**

В Pandas корреляцию можно реализовать с помощью метода
.corr(). Он позволяет вычислить корреляцию между всеми парами
столбцов в датафрейме или между двумя конкретными столбцами.
Коэффициент корреляции может находиться в диапазоне от -1 до
1, где -1 означает полную отрицательную корреляцию, 1 -
положительную, а 0 - отсутствие корреляции.
Метод .corr() возвращает матрицу корреляции для всех числовых
столбцов в датафрейме.

```python
df3 = pd.DataFrame({'Средний балл':[4.5,3.2,4.8,3.9,4.2], 'Часы учебы в неделю':[20,15,25,18,22], 'Количество сессий':[2,3,1,2,2]})
df3
```

```python
df3.corr()
```
